$y_i = \alpha + \beta*x_i + e_i  = 10 + 2*x_i + e_i  $

$x_i = e_i - 0.9 * e_{i-1}$

```r
e_2 = rnorm(n = 101, 0,1)
e = e_2[2:101]
x = e - 0.9*e_2[1:100] 
y = 10 + 2 * x + e 
```

```r
mod<-lm(y~x)
summary(mod)[[4]][2,1]
```

```r
betas_1 <- c()
for (i in 1:500){
    e = rnorm(n = 100, 0,1)
    y = 10 + 2 * x + e 
    mod<-lm(y~x) 
    betas_1 <- c(betas_1, summary(mod)[[4]][2,1])
}
```

```r
hist(betas_1)
mean(betas_1)
```
